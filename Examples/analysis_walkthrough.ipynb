# Dims<br>

# Switch <br>
 DIMS Objects can switch x, y, hue, col, row by changing the dims object either inplace or by generating a copy"

```python
dims = Dims(**dict(y="signal", x="timepoint", hue="event", col="subject") )
```

```python
print("#\t 'dims.by' before switching:", dims.by,)
print("#\t 'dims.by' in chain:        ", dims.switch(col = "event", inplace=False).by)
print("#\t 'dims.by' after switching: ", dims.by, "(inplace=False)")
print("#\t 'dims.by' in chain:        ", dims.switch(col = "event", inplace=True).by)
print("#\t 'dims.by' after switching: ", dims.by, "(inplace=True)")
```

 Switching method can be called passing in two *args without knowing the column names. One of them may be one of `[y,x,hue,row,col]`, which is helpful if you don't remember the exact column names "

# Analysis<br>
<br>
##### Analysis class is composed by:<br>
- **Dims object**<br>
- **Pandas Dataframe** for Raw Data<br>
- **Filer Class** Is automatically set. Determines output path (and input path if pyrectories is used)<br>
- **Project Title** `(str)` to use as figure suptitles and filenames for outputs<br>
- **properties:**<br>
    - factors_all: Stores all factors `{x, y, hue, col, row}`<br>
- **methods**<br>
    - getfactors(): Retrieves the column name when one of `{x, y, hue, col, row}` is passed<br>
    - switch(), set() is callable on Analysis object (same funitonality as dims)

```python
print("\n### ANALYSIS Objects can also switch x, y, hue, col, row by changing the dims object either inplace or by generating a copy")
```

```python
df, _ = ut.load_dataset("fmri") # _ stores a dictionary with dims. For clarity, we initialize analysis explicitly
ana = ut.Analysis(dims = dict(y="signal", x="timepoint", hue="event", col="subject"),
                  title= "APROJECT", data=df)
```

```python
print("#\t 'ana.dims.by' before switching:", ana.dims.by,)
print("#\t 'ana.dims.by' in chain:        ", ana.switch(col = "event", inplace=False).dims.by)
print("#\t 'ana.dims.by' after switching: ", ana.dims.by, "(inplace=False)")
print("#\t 'ana.dims.by'  in chain:       ", ana.switch(col = "event", inplace=True).dims.by)
print("#\t 'ana.dims.by' after switching: ", ana.dims.by, "(inplace=True)")
```
