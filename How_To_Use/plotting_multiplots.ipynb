# MultiPlots
MultiPlots are plots that
- Combine multiple layers of different `seaborn` plotting functions
- Use a pre-defined set of parameters optimized for that plotting combination
- Allow for customization of each layer 

### Import plotastic and example Data

```python
import plotastic as plst

# Import Example Data
DF, _dims = plst.load_dataset("fmri", verbose = False)

DF.head()
```

### Make DataAnalysis Object

```python
dims = dict(
    y= 'signal', 
    x= 'timepoint',
    hue= 'event', 
    col= 'region',
)

DA = plst.DataAnalysis(
    data=DF,           # Dataframe
    dims=dims,         # Dictionary with y, x, hue, col, row 
    subject="subject", # Data is paired by subject (optional)
    verbose=True,      # Print out Info about the Data
)
```

## Examples
And how to edit what's inside the axes

#### Boxplot + Stripplot (default)

```python
DA.plot_box_strip()
```

#### Boxplot + Stripplot (customized)

```python
DA.plot_box_strip(
    ### General keywords for faster access
    marker_size = .9,     # Size of the stripplot markers
    marker_alpha = .9,  # Alpha of the stripplot markers
    legend=True,

    ### Subplot keywords:
    subplot_kws = dict(
        figsize=(10, 3), # Figure size
        sharey=False,   # Each Axes scales y-axis individually
        wspace=.2,      # Horizontal space between subplots
        ),

    
    ### Boxplot keywords:
    box_kws = dict(
        palette="Paired",
        showfliers=False,
        # Use boxprops, whiskerprops, capprops, and flierprops
        capprops=dict(  #' Caps at the end of whiskers
            alpha=.9,
            zorder=200,
            linewidth=3,
            ),
        ),
    
    ### Stripplot Keywords
    strip_kws = dict(
        palette="Paired",
        dodge=True,     # Separates the points in hue
        jitter=0.2,     # How far datapoints of one group scatter across the x-axis
        zorder=100,
        ### Marker Style
        size=3,       # Overwrite marker_size
        alpha=.5,      # Overwrite marker_alpha
        edgecolor="white",
        linewidth=.5,   # Edge width of the marker
        ),
    
    ### Legend Keywords
    legend_kws = dict(
        bbox_to_anchor=(1.05, 0.5), # TODO: Legend positioning is kinda painful
        # loc=2, 
        # pad=-0.5,
        ),
    )

```

#### Boxplot + Swarmplot (default)
Swarmplot makes sure that the points don't overlap. Hence, pointsize is
smaller by default 

```python
DA.plot_box_strip()
```

```python
DA.plot_box_swarm(
    ### General keywords for faster access
    marker_size = .9,     # Size of the stripplot markers
    marker_alpha = .5,  # Alpha of the stripplot markers
    legend=True,

    ### Subplot keywords:
    subplot_kws = dict(
        figsize=(10, 3), # Figure size
        sharey=False,   # Each Axes scales y-axis individually
        wspace=.2,      # Horizontal space between subplots
        ),

    
    ### Boxplot keywords:
    box_kws = dict(
        palette="Paired",
        showfliers=False,
        # Use boxprops, whiskerprops, capprops, and flierprops
        capprops=dict(  #' Caps at the end of whiskers
            alpha=.9,
            zorder=200,
            linewidth=3,
            ),
        ),
    
    ### Stripplot Keywords
    swarm_kws = dict(
        palette="Paired",
        dodge=True,     # Separates the points in hue
        zorder=100,
        ### Marker Style
        size=2,       # Overwrite marker_size
        alpha=1,      # Overwrite marker_alpha
        edgecolor="white",
        linewidth=.1,   # Edge width of the marker
        ),
    
    ### Legend Keywords
    legend_kws = dict(
        bbox_to_anchor=(1.05, 0.5), # TODO: Legend positioning is kinda painful
        # loc=2, 
        # pad=-0.5,
        ),
    )
```
